### Model Development and Refinement

**Segragating the data**

```python
# Divide data to inputs and outputs
X=data.iloc[:,:-1].values
y=data.iloc[:,1].values
```

**Splitting the dataset to train and test set**

```python
from sklearn.model_selection import train_test_split  
X_train, X_test, y_train, y_test = train_test_split(X, y,test_size=0.2, random_state=0)
```

### Training our algorithm

**The Algorithm we are using is - Linear Regression**

```python
from sklearn.linear_model import LinearRegression  
model = LinearRegression()  
model.fit(X_train, y_train) 
```

#### We know that a linear equation is of type: y=mx+c
we incorporate this during our plot

```python
print('Intercept value is :',model.intercept_)
print('Linear coefficient is:',model.coef_)5
```

```python
# Plotting the regression line
line = model.coef_*X + model.intercept_

# Plotting for the test data
plt.scatter(X,y)
plt.plot(X, line);
plt.show()
```

### Making Predictions

```python
# Predicting the scores
y_pred = model.predict(X_test) 
```

### Comparing Actual vs Predicted

```python
df = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred})  
df
```

```python
#plotting the grid to depict the actual and predicted value

df.plot(kind='bar',figsize=(5,5), color=('blue','green'))
plt.grid(which='major', linewidth='0.5', color='green')
plt.grid(which='minor', linewidth='0.5', color='black')
plt.show()
```

### Checking the error

```python
from sklearn import metrics  
print('Mean Absolute Error:', 
      metrics.mean_absolute_error(y_test, y_pred))
```

## Accuracy

```python
print('R^2:', metrics.r2_score(y_test, y_pred))
```
